````python
import pyinputplus as pyip
def add(a, b):
    return a + b

def subtract(a, b):
    return a - b

def multiply(a, b):
    return a * b

def divide(a, b):
    if b == 0:
        return "Error: Cannot divide by zero."
    else:
        return a / b

def get_operation():
    print("Simple Calculator Menu")
    print("1. Add")
    print("2. Subtract")
    print("3. Multiply")
    print("4. Divide")

    choice = pyip.inputChoice(['1','2','3','4'], prompt="Enter your choice (1-4): ")
    return choice

def get_number(prompt):
    return pyip.inputFloat(prompt)

def perform_calculation(operation, num1, num2):
    if operation == "1":
        return add(num1, num2)
    elif operation == "2":
        return subtract(num1, num2)
    elif operation == "3":
        return multiply(num1, num2)
    elif operation == "4":
        return divide(num1, num2)
    else:
        return "Invalid choice."

while True:
    operation = get_operation()
    num1 = get_number("Enter the first number: ")
    num2 = get_number("Enter the second number: ")

    result = perform_calculation(operation, num1, num2)

    print("Result:", result)

    again = input("Do you want to calculate again? (yes/no): ")
    if again.lower() != "yes":
        print("Goodbye!")
        break
````
