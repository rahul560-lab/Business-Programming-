````python
class BudgetFinanceManager(SimpleFinanceManager):
    def __init__(self, monthly_budget):
        super().__init__()
        self.monthly_budget = float(monthly_budget)

    def get_budget_remaining(self):
        total_expenses = abs(self.get_total_expenses())
        return self.monthly_budget - total_expenses

    def get_budget_status(self):
        remaining = self.get_budget_remaining()

        if remaining >= 0:
            return (
                f"Monthly Budget: ${self.monthly_budget:.2f}\n"
                f"Total Spending: ${abs(self.get_total_expenses()):.2f}\n"
                f"Budget Remaining: ${remaining:.2f}\n"
                f"Status: You are within your budget."
            )
        else:
            return (
                f"Monthly Budget: ${self.monthly_budget:.2f}\n"
                f"Total Spending: ${abs(self.get_total_expenses()):.2f}\n"
                f"Over Budget By: ${abs(remaining):.2f}\n"
                f"Status: You have gone over your budget."
            )
````

````python
!pip install gradio pandas -q

import gradio as gr
import pandas as pd
from datetime import datetime


class Transaction:
    def __init__(self, description, amount, category, date=None):
        self.description = description
        self.amount = float(amount)
        self.category = category

        if date is None:
            self.date = datetime.today().strftime("%Y-%m-%d")
        else:
            self.date = date

    def is_expense(self):
        return self.amount < 0

    def transaction_type(self):
        if self.is_expense():
            return "Expense"
        else:
            return "Income"

    def __str__(self):
        return f"{self.date} | {self.description} | ${self.amount:.2f} | {self.category} | {self.transaction_type()}"


class SimpleFinanceManager:
    def __init__(self):
        self.transactions = []

    def add_transaction(self, transaction):
        self.transactions.append(transaction)

    def get_total_expenses(self):
        total = 0
        for transaction in self.transactions:
            if transaction.is_expense():
                total += transaction.amount
        return total

    def get_total_income(self):
        total = 0
        for transaction in self.transactions:
            if not transaction.is_expense():
                total += transaction.amount
        return total

    def get_balance(self):
        return self.get_total_income() + self.get_total_expenses()

    def get_summary(self):
        return (
            f"Total Income: ${self.get_total_income():.2f}\n"
            f"Total Expenses: ${self.get_total_expenses():.2f}\n"
            f"Balance: ${self.get_balance():.2f}"
        )

    def get_all_transactions(self):
        if len(self.transactions) == 0:
            return "No transactions added yet."

        text = ""
        for transaction in self.transactions:
            text += str(transaction) + "\n"

        return text


app_manager = SimpleFinanceManager()


def add_transaction_via_gradio(description, amount, category):
    if description is None or description.strip() == "":
        return "Please enter a description.", app_manager.get_summary(), app_manager.get_all_transactions()

    transaction = Transaction(description, amount, category)
    app_manager.add_transaction(transaction)

    return (
        f"Transaction added successfully:\n{transaction}",
        app_manager.get_summary(),
        app_manager.get_all_transactions()
    )


def load_transactions_from_csv(csv_file):
    if csv_file is None:
        return "No file uploaded.", app_manager.get_summary(), app_manager.get_all_transactions()

    try:
        df = pd.read_csv(csv_file.name)

        for index, row in df.iterrows():
            transaction = Transaction(
                description=row["description"],
                amount=row["amount"],
                category=row["category"],
                date=row["date"]
            )
            app_manager.add_transaction(transaction)

        return (
            "CSV transactions loaded successfully.",
            app_manager.get_summary(),
            app_manager.get_all_transactions()
        )

    except Exception as e:
        return f"Error loading CSV: {str(e)}", app_manager.get_summary(), app_manager.get_all_transactions()


def refresh_analysis():
    return app_manager.get_summary(), app_manager.get_all_transactions()


def reset_app():
    global app_manager
    app_manager = SimpleFinanceManager()
    return "App reset successfully.", app_manager.get_summary(), app_manager.get_all_transactions()


with gr.Blocks(title="Smart Finance App v2") as enhanced_demo:

    gr.Markdown("# Smart Finance App v2")
    gr.Markdown("Add transactions manually or load transactions from a CSV file.")

    with gr.Tab("Add Transactions"):
        desc_input = gr.Textbox(label="Description", placeholder="Example: Coffee Shop")
        amount_input = gr.Number(
            label="Amount",
            value=0.0,
            info="Use negative numbers for expenses and positive numbers for income"
        )
        category_input = gr.Dropdown(
            label="Category",
            choices=["Food", "Transport", "Entertainment", "Bills", "Income", "Other"],
            value="Food"
        )

        add_button = gr.Button("Add Transaction")
        status_output = gr.Textbox(label="Status", interactive=False)

    with gr.Tab("Load from CSV"):
        csv_upload = gr.File(label="Upload CSV File")
        load_button = gr.Button("Load CSV")
        csv_status = gr.Textbox(label="CSV Status", interactive=False)

    with gr.Tab("Analysis"):
        summary_display = gr.Textbox(
            label="Summary",
            value=app_manager.get_summary(),
            lines=4,
            interactive=False
        )

        all_transactions = gr.Textbox(
            label="All Transactions",
            value=app_manager.get_all_transactions(),
            lines=8,
            interactive=False
        )

        refresh_button = gr.Button("Refresh Analysis")
        reset_button = gr.Button("Reset App")

    add_button.click(
        fn=add_transaction_via_gradio,
        inputs=[desc_input, amount_input, category_input],
        outputs=[status_output, summary_display, all_transactions]
    )

    load_button.click(
        fn=load_transactions_from_csv,
        inputs=csv_upload,
        outputs=[csv_status, summary_display, all_transactions]
    )

    refresh_button.click(
        fn=refresh_analysis,
        inputs=[],
        outputs=[summary_display, all_transactions]
    )

    reset_button.click(
        fn=reset_app,
        inputs=[],
        outputs=[status_output, summary_display, all_transactions]
    )


enhanced_demo.launch(share=True, debug=True)
````

````python
class SimpleFinanceManager:
    def __init__(self):
        self.transactions = []

    def add_transaction(self, transaction):
        self.transactions.append(transaction)

    def get_total_expenses(self):
        total = 0
        for transaction in self.transactions:
            if transaction.is_expense():
                total += transaction.amount
        return total

    def get_spending_by_category(self, category):
        total = 0
        for transaction in self.transactions:
            if transaction.category == category and transaction.is_expense():
                total += transaction.amount
        return total

    def get_recent_transactions(self, count=5):
        return self.transactions[-count:]

    def get_summary(self):
        total_expenses = self.get_total_expenses()
        total_income = 0

        for transaction in self.transactions:
            if not transaction.is_expense():
                total_income += transaction.amount

        balance = total_income + total_expenses

        return f"Total Income: ${total_income:.2f}\nTotal Expenses: ${total_expenses:.2f}\nBalance: ${balance:.2f}"


manager = SimpleFinanceManager()

manager.add_transaction(Transaction("Coffee Shop", -4.50, "Food"))
manager.add_transaction(Transaction("Bus Ticket", -3.20, "Transport"))
manager.add_transaction(Transaction("Salary", 500.00, "Income"))
manager.add_transaction(Transaction("Groceries", -67.80, "Food"))

print(manager.get_summary())

print("\nFood spending:")
print(manager.get_spending_by_category("Food"))

print("\nRecent transactions:")
for transaction in manager.get_recent_transactions():
    print(transaction)
````

````python
class Transaction:
    def __init__(self, description, amount, category, date=None):
        self.description = description
        self.amount = float(amount)
        self.category = category

        if date is None:
            self.date = datetime.today().strftime("%Y-%m-%d")
        else:
            self.date = date

    def is_expense(self):
        return self.amount < 0

    def __str__(self):
        if self.is_expense():
            transaction_type = "Expense"
        else:
            transaction_type = "Income"

        return f"{self.date} | {self.description} | ${self.amount:.2f} | {self.category} | {transaction_type}"

test_transaction = Transaction("Coffee", -4.50, "Food")
print(test_transaction)
print(f"Is expense: {test_transaction.is_expense()}")
````
