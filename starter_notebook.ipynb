### Foundation Data Processing Functions

````python
import pandas as pd

# 🤖 AI Collaboration: Data Loading and Cleaning Function
# Ask AI to help you create a robust data loading function

def load_and_clean_transaction_data(data_input):
    """
    Load and clean transaction data for the Smart Finance Assistant.
    This function can accept either a file path to a CSV or a pandas DataFrame.

    🤖 AI Collaboration Prompt:
    "Create a function to load CSV transaction data with Date, Amount, Category,
    Description columns. Handle dollar signs in Amount, missing values, and
    data validation. Include clear business-focused error messages."

    Args:
        data_input: Path to CSV file or a pandas DataFrame.

    Returns:
        pandas.DataFrame: Cleaned transaction data, or None if an error occurs.
    """
    df = None
    if isinstance(data_input, str):
        # Assume it's a file path
        if not data_input.endswith('.csv'):
            print("❌ Error: Input file must be a CSV.")
            return None
        try:
            df = pd.read_csv(data_input)
        except FileNotFoundError:
            print(f"❌ Error: File not found at {data_input}")
            return None
        except Exception as e:
            print(f"❌ Error reading CSV file: {e}")
            return None
    elif isinstance(data_input, pd.DataFrame):
        df = data_input.copy() # Work on a copy to avoid modifying original
    else:
        print("❌ Error: Invalid input type. Please provide a file path or a pandas DataFrame.")
        return None

    if df is None or df.empty:
        print("⚠️ Warning: No data to process or DataFrame is empty.")
        return None

    # Check for required columns
    required_columns = ['Date', 'Amount', 'Category', 'Description']
    if not all(col in df.columns for col in required_columns):
        missing_cols = [col for col in required_columns if col not in df.columns]
        print(f"❌ Error: Missing required columns: {', '.join(missing_cols)}.")
        return None

    # Clean Amount column
    # Convert to string first to handle mixed types safely, then remove '$', and convert to numeric
    df['Amount'] = df['Amount'].astype(str).str.replace('$', '', regex=False)
    df['Amount'] = pd.to_numeric(df['Amount'], errors='coerce')

    # Convert Date column to datetime, coercing invalid dates to NaT (Not a Time)
    df['Date'] = pd.to_datetime(df['Date'], errors='coerce')

    # Drop rows where 'Amount' or 'Date' are invalid (NaN/NaT)
    original_rows = len(df)
    df.dropna(subset=['Amount', 'Date'], inplace=True)
    if len(df) < original_rows:
        print(f"ℹ️ {original_rows - len(df)} rows dropped due to invalid 'Amount' or 'Date' values.")

    if df.empty:
        print("⚠️ Warning: DataFrame is empty after cleaning. No valid transactions found.")
        return None

    print("✅ Transaction data loaded and cleaned successfully.")
    return df

# Test your function with df_sample now
# cleaned_df = load_and_clean_transaction_data(df_sample)
# if cleaned_df is not None:
#     print(cleaned_df.head())
````

````python
# 🤖 AI Collaboration: Spending Analysis Function
# Ask AI to help you create comprehensive spending analysis

def analyze_spending_patterns(df):
    """
    Analyze spending patterns and generate business insights

    🤖 AI Collaboration Prompt:
    "Create a function that analyzes spending by category, calculates percentages,
    identifies top spending areas, and generates actionable financial insights
    formatted for business presentation."

    Args:
        df: Cleaned transaction DataFrame
    Returns:
        dict: Analysis results and insights
    """
    if df.empty:
        return {
            "message": "No transactions to analyze.",
            "total_spending": "$0.00", # Changed to formatted string for consistency
            "spending_by_category": {},
            "insights": []
        }

    # Calculate total spending (excluding refunds, or treating them as negative spending)
    total_spending = df['Amount'].sum()

    # Spending by category
    spending_by_category = df.groupby('Category')['Amount'].sum().sort_values(ascending=False)

    # Calculate percentage of total spending for each category
    spending_percentages = (spending_by_category / spending_by_category.sum()) * 100

    # Prepare insights
    insights = []
    if not spending_by_category.empty:
        top_category = spending_by_category.index[0]
        top_category_amount = spending_by_category.iloc[0]
        insights.append(f"Your largest spending category is '{top_category}' with a total of ${top_category_amount:,.2f}.")

        if total_spending > 0:
            insights.append(f"Total spending across all categories is ${total_spending:,.2f}.")

            # Identify potential savings if negative amounts are significant (refunds)
            positive_spending = df[df['Amount'] > 0]['Amount'].sum()
            negative_spending = df[df['Amount'] < 0]['Amount'].sum()
            if negative_spending < -0.01: # Check if there are significant refunds
                insights.append(f"You have received refunds totaling ${abs(negative_spending):,.2f}, which is great for managing expenses!")

    # Format output for presentation
    formatted_spending_by_category = spending_by_category.apply(lambda x: f"${x:,.2f}").to_dict()
    formatted_spending_percentages = spending_percentages.apply(lambda x: f"{x:,.2f}%").to_dict()

    return {
        "message": "Spending analysis completed successfully.",
        "total_spending": f"${total_spending:,.2f}",
        "spending_by_category": formatted_spending_by_category,
        "spending_percentages": formatted_spending_percentages,
        "insights": insights
    }

# Test your function
# analysis = analyze_spending_patterns(clean_data)
# print(analysis)
````

## Custom Financial Tools

````python
from hands_on_ai import agent

def create_savings_calculator_tool():
    """
    Create a custom savings goal calculator as an agent tool

    ♐ AI Collaboration Prompt:
    "Create a savings goal calculator function that takes current savings,
    monthly contribution, and target amount, then calculates time to reach
    goal. Format output for user-friendly display."
    """
    def savings_calculator(current_savings: float, monthly_contribution: float, target_amount: float) -> str:
        """
        Calculates the time (in months and years) required to reach a savings goal.
        """
        if monthly_contribution <= 0:
            return "Monthly contribution must be positive to reach a savings goal."
        if target_amount <= current_savings:
            return "Your current savings already meet or exceed your target amount!"

        remaining_amount = target_amount - current_savings
        months_to_reach_goal = remaining_amount / monthly_contribution

        years = int(months_to_reach_goal // 12)
        months = int(months_to_reach_goal % 12)

        return (f"You will reach your savings goal of ${target_amount:,.2f} in "
                f"{years} years and {months} months, "
                f"contributing ${monthly_contribution:,.2f} monthly.")

    return savings_calculator

# Register your tool with the agent system
# You would typically register the *instance* of the function returned by create_savings_calculator_tool
savings_tool = create_savings_calculator_tool()
# Assuming 'agent.register_tool' is available and works similarly to other tool registration systems.
# The actual hands-on-ai agent tool registration mechanism might differ based on the library version.
# For now, we'll just create the function and comment out the registration until 'hands_on_ai.agent' is fully explored.
agent.register_tool(name="savings_calculator", function=savings_tool, description="Calculates the time required to reach a savings goal based on current savings, monthly contributions, and a target amount.")
print("✅ Savings Calculator tool registered with hands-on-ai agent system.")

# Test the savings calculator
print(savings_tool(current_savings=100.0, monthly_contribution=50.0, target_amount=1000.0))
print(savings_tool(current_savings=1000.0, monthly_contribution=100.0, target_amount=900.0))
print(savings_tool(current_savings=0.0, monthly_contribution=0.0, target_amount=500.0))
````

````python
import gradio as gr
import pandas as pd

# Assuming these functions are defined and available globally or imported
# from your notebook context (e.g., load_and_clean_transaction_data, analyze_spending_patterns, generate_financial_recommendations, finny_system_prompt, savings_tool)

# Import get_response from hands_on_ai.chat (already done in GPH4sGKN6WsN, but explicit here for clarity in this cell)
from hands_on_ai.chat import get_response

def create_finance_assistant_ui():
    """
    Create a comprehensive Gradio interface for the Smart Finance Assistant

    → AI Collaboration Prompt:
    "Help me design a Gradio interface that combines CSV upload, spending
    analysis, chat functionality, and custom tools in a user-friendly
    layout suitable for a personal finance application."
    """

    with gr.Blocks(title="Smart Finance Assistant") as demo:
        gr.Markdown("# 🏦 Smart Finance Assistant")
        gr.Markdown("Your personal AI-powered finance manager.")

        with gr.Tab("Data Analysis"): # Tab for CSV upload and analysis
            gr.Markdown("### Upload your Transaction Data (CSV)")
            file_upload = gr.File(label="Upload CSV", file_types=[".csv"])
            analyze_button = gr.Button("Analyze Spending")

            # Output areas for data analysis
            gr.Markdown("### Spending Analysis Results")
            analysis_output = gr.JSON(label="Analysis Summary")
            recommendations_output = gr.Markdown(label="Financial Recommendations")

            # Function to handle CSV upload and analysis
            def process_and_analyze(file):
                if file is None:
                    return {}, "Please upload a CSV file."

                df = pd.read_csv(file.name)
                cleaned_df = load_and_clean_transaction_data(df)

                if cleaned_df is None or cleaned_df.empty:
                    return {}, "No valid data to analyze after cleaning."

                analysis_results_dict = analyze_spending_patterns(cleaned_df)
                recommendations_str = generate_financial_recommendations(analysis_results_dict)

                return analysis_results_dict, recommendations_str

            analyze_button.click(
                process_and_analyze,
                inputs=file_upload,
                outputs=[analysis_output, recommendations_output]
            )

        with gr.Tab("AI Chat (Finny)"): # Tab for AI Chat
            gr.Markdown("### Chat with Finny, your Financial Advisor")
            chatbot = gr.Chatbot(label="Finny Chat")
            msg = gr.Textbox(label="Your Message")
            clear = gr.ClearButton([msg, chatbot])

            # Modified chat response logic using hands_on_ai.chat.get_response
            def respond_to_chat(message, chat_history):
                # Use the finny_system_prompt defined globally
                global finny_system_prompt
                response = get_response(message, system_prompt=finny_system_prompt)
                chat_history.append((message, response))
                return "", chat_history

            msg.submit(respond_to_chat, [msg, chatbot], [msg, chatbot])

        with gr.Tab("Financial Tools"): # Tab for custom tools
            gr.Markdown("### Use our Financial Calculators")
            gr.Markdown("#### Savings Goal Calculator")
            current_s = gr.Number(label="Current Savings ($")
            monthly_c = gr.Number(label="Monthly Contribution ($")
            target_a = gr.Number(label="Target Amount ($")
            calc_button = gr.Button("Calculate Time to Goal")
            savings_output = gr.Textbox(label="Time to Reach Goal")

            # Link with the savings_tool function
            calc_button.click(
                savings_tool,
                inputs=[current_s, monthly_c, target_a],
                outputs=savings_output
            )

    return demo

# Launch your complete application
demo = create_finance_assistant_ui()
demo.launch()
````
